```python
# Predefined stock prices
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 2700,
    "AMZN": 3300,
    "MSFT": 300
}
```

```python
portfolio = {}

print("Enter your stocks (type 'done' to stop):")

while True:
    stock = input("Stock name: ").upper()
    
    if stock == "DONE":
        break
    
    if stock not in stock_prices:
        print("This stock is not available.")
        continue
    
    qty = int(input("Quantity: "))
    
    # Add or update stock
    if stock in portfolio:
        portfolio[stock] += qty
    else:
        portfolio[stock] = qty
```

```python
total = 0

print("\nYour Portfolio:")
print("------------------")

for stock in portfolio:
    price = stock_prices[stock]
    qty = portfolio[stock]
    value = price * qty
    
    total += value
    
    print(stock, ":", qty, "shares =", value)

print("------------------")
print("Total Investment =", total)
```

```python
with open("portfolio.txt", "w") as file:
    file.write("Portfolio Summary\n")
    
    for stock in portfolio:
        price = stock_prices[stock]
        qty = portfolio[stock]
        value = price * qty
        
        file.write(f"{stock}: {qty} shares = {value}\n")
    
    file.write(f"\nTotal = {total}")

print("Saved to file successfully!")
```
